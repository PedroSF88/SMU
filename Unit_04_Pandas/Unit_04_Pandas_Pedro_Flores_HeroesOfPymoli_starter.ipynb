```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# total number of player
playercount=len(purchase_data.SN.value_counts())
playercount
```

```python
# unique items
uniqueitems=len(purchase_data['Item ID'].unique())
# number of purchases
purchases=purchase_data.SN.count()
#average price of items
avgprice=purchase_data.Price.mean()
#total revenue
totalrev=purchase_data['Price'].sum()
```

```python
# list of each variable
uniqueitemscols=[uniqueitems]
purchasescols=[purchases]
avgpricecols=[avgprice]
totalrevcols=[totalrev]
```

```python
#empty data frame
summarydf= pd.DataFrame()
```

```python
#adding each variable into the empty data frame
summarydf["Number of Unique Items"]=uniqueitemscols
summarydf["Average Price"]=avgpricecols
summarydf["Number of Purchases"]=purchasescols
summarydf["Total Revenue"]=totalrev
summarydf['Total Revenue']=summarydf['Total Revenue'].map("${:.2f}".format)
summarydf['Average Price']=summarydf['Average Price'].map("${:.2f}".format)
summarydf


```

```python
# Count number of each gender
malesn=purchase_data.loc[purchase_data['Gender']=='Male']
maleunique=malesn.SN.nunique()
femalesn=purchase_data.loc[purchase_data['Gender']=='Female']
femaleunique=femalesn.SN.nunique()
othersn=purchase_data.loc[purchase_data['Gender']=='Other / Non-Disclosed']
otherunique=othersn.SN.nunique()
```

```python
#new data frame
percentagesdf=pd.DataFrame()
```

```python
# naming gender and num of gender
gender=['Male','Female','Other/ Non-Disclosed']
data=[maleunique,femaleunique,otherunique]
```

```python
#combine gender w num of gender
percentagesdf=pd.DataFrame(data)
percentagesdf.index=gender
percentagesdf.columns=['Total Counts']

```

```python
# adding percentages to df
percentagesdf['Percentage']=percentagesdf['Total Counts']/percentagesdf['Total Counts'].sum()*100
percentagesdf['Percentage']=percentagesdf['Percentage'].map("%{:.2f}".format)
percentagesdf
```

```python
#new data frame
purchase_data2=pd.DataFrame()
```

```python
#group by gender to run functions
purchase_data2['Purchase Count']=purchase_data.groupby(['Gender'])['Item ID'].count()
purchase_data2['Average Purchase Price']=purchase_data.groupby(['Gender'])['Price'].mean()
purchase_data2['Total Purchase Value']=purchase_data.groupby(['Gender'])['Price'].sum()
purchase_data2['Average per Person']=purchase_data2['Total Purchase Value']/purchase_data.groupby(['Gender'])['SN'].nunique()
```

```python
# copy df
df=purchase_data
```

```python
#dropping duplicates
dfdropped= df.drop_duplicates("SN",keep='first')
```

```python
#new df to add 
age_data=pd.DataFrame()
```

```python
# Bins  
bins = [0, 9, 14, 19, 24, 29,34, 39,100]

# Create the names for the  bins
names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
ages = pd.cut(dfdropped["Age"], bins, labels=names)
data=ages.value_counts()
age_data["Total Count"]=data
age_data["Percentage"]=data/playercount*100
age_data["Percentage"]=age_data["Percentage"].map("%{:.2f}".format)

age_data
```

```python
#new df
age_set=pd.DataFrame()
```
